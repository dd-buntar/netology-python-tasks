## Задача 1

Напишите функцию, которая возвращает название валюты (поле ‘Name’) с максимальным значением курса с помощью сервиса https://www.cbr-xml-daily.ru/daily_json.js

```python
import requests
```

```python
def get_currency_with_max_rate():
    response = requests.get('https://www.cbr-xml-daily.ru/daily_json.js')
    data = response.json()
    valute = data['Valute']
    
    max_currency = ''
    max_value = 0

    for code in valute:
        if valute[code]['Value'] > max_value:
            max_value = valute[code]['Value']
            max_currency = valute[code]['Name']
    
    return max_currency
```

## Задача 2

Добавьте в класс `Rate` параметр `diff` (со значениями `True` или `False`), который в случае значения `True` в методах курсов валют (`eur`, `usd` и т.д.) будет возвращать не курс валюты, а изменение по сравнению в прошлым значением. Считайте, `self.diff` будет принимать значение `True` только при возврате значения курса. При отображении всей информации о валюте он не используется.

```python
class Rate:
    def __init__(self, format_='value', diff=False):
        self.format = format_
        self.diff = diff
    
    def exchange_rates(self):
        r = requests.get('https://www.cbr-xml-daily.ru/daily_json.js')
        return r.json()['Valute']
    
    def make_format(self, currency):
        response = self.exchange_rates()
        
        if currency in response:
            if self.format == 'full':
                return response[currency]
            
            if self.format == 'value':
                if self.diff:
                    current = response[currency]['Value']
                    previous = response[currency]['Previous']
                    return round(current - previous, 4)
                else:
                    return response[currency]['Value']
        
        return 'Error'
    
    def eur(self):
        return self.make_format('EUR')
    
    def usd(self):
        return self.make_format('USD')
    
    def azn(self):
        return self.make_format('AZN')
    
    def brl(self):
        return self.make_format('BRL')
```

```python
print('Полная информация:')
r3 = Rate(format_='full', diff=True)
print(r3.usd()['Name'], '-', r3.usd()['Value'])
```

## Задача 3

Напишите класс `Designer`, который учитывает количество международных премий.  
Подсказки в коде занятия (“Ноутбук к лекциям «Понятие класса» + презентация”, zip-файл “Используемый ноутбук к лекциям «Понятие класса»).

```python
from python_class_example import Employee
```

```python
class Designer(Employee):
    def __init__(self, name, seniority, awards=2):
        super().__init__(name, seniority)
        self.awards = awards
        self.start_points = awards * 2 
    
    def check_if_it_is_time_for_upgrade(self):
        self.seniority += 1
        
        total_points = self.seniority + self.start_points
        
        if total_points % 7 == 0:
            self.grade_up()
        
        return self.publish_grade()
```

```python
anna = Designer('Анна', 0, awards=2)

for i in range(20):
    anna.check_if_it_is_time_for_upgrade()
```

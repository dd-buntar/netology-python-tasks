## Задание 1
Датасет для домашнего задания находится в материалах занятия - файлы для домашнего задания «Библиотека Pandas».
Определите, какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('data_task1/ratings.csv')
movies = pd.read_csv('data_task1/movies.csv')

merged = ratings.merge(movies, on='movieId', how='left')
rating_5 = merged[merged['rating'] == 5.0]
rating_5_count = rating_5.groupby('title').size().reset_index(name='count_5')
top_movie = rating_5_count.loc[rating_5_count['count_5'].idxmax()]
```

```python
print(f"Фильм с наибольшим количеством оценок 5.0: {top_movie['title']}, количество - {top_movie['count_5']}")
```

## Задание 2
По данным файла `power.csv` посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. Не учитывайте в расчётах отрицательные значения `quantity`.

```python
df = pd.read_csv('data_task2/power.csv')

baltic_countries = ['Lithuania', 'Latvia', 'Estonia']

mask = (df['country'].isin(baltic_countries)) & \
       (df['category'].isin([4, 12, 21])) & \
       (df['year'].between(2005, 2010)) & \
       (df['quantity'] > 0)

result = df[mask]['quantity'].sum()
print(f"Суммарное потребление стран Прибалтики: {result}")
```
